```python
import rdkit
from rdkit import Chem
print(rdkit.__version__)
```

```python
smiles = "C[C@H]1CCC[N@@H+]([C][C][C](C(=O)C([O])=O)[C@]2([O])[C]3[C][C]c4oc(Oc5cccc([C@]6(c7nncn7C)C[C@@H](C)C6)c5)c([C]([C][N])[C]=O)c4[C]32)C1"
mol = Chem.MolFromSmiles(smiles)
mol
```

```python
Chem.MolFromSmiles(Chem.MolToSmiles(mol))
```

`Elements in brackets` represent `Radical Electrons`, which have unpaired electrons.

```python
def neutralizeRadicals(mol):
    for a in mol.GetAtoms():
        if a.GetNumRadicalElectrons()>0:
            ## in-place
            a.SetNumRadicalElectrons(0)
```

```python
neutralizeRadicals(mol)
print(f"{Chem.MolToSmiles(mol)}")
mol
```

```python
Chem.MolFromSmiles(Chem.MolToSmiles(mol))
```
