```python
import rdkit
from rdkit import Chem
print(rdkit.__version__)
```

```python
smiles = "CN(C)C/C=C/C(=O)NC1=C(C=C2C(=C1)C(=NC=N2)NC3=CC(=C(C=C3)F)Cl)O[C@H]4CCOC4"
mol = Chem.MolFromSmiles(smiles)
mol
```

## set `AtomMapNum` in mol

```python
def annotating_atoms(mol):
    [atom.SetAtomMapNum(_+1) for _, atom in enumerate(mol.GetAtoms())]
```

```python
annotating_atoms(mol)
mol
```

## remove `AtomMapNum` in mol

```python
def clear_annotating_atoms(mol):
    [atom.SetAtomMapNum(0) for _, atom in enumerate(mol.GetAtoms())]
```

```python
clear_annotating_atoms(mol)
mol
```
